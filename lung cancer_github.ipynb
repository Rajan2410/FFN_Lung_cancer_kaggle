```python
x=data.drop(['Survival_Months'],axis=1)
```

```python
y=data['Survival_Months']
```

```python
y=np.log(y)
```

```python
from sklearn.model_selection import RandomizedSearchCV
from sklearn.metrics import mean_absolute_error
import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
from scipy.stats import skew

skewness=skew(y_train)

skewness
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
std=StandardScaler()
```

```python
x_train_std=std.fit_transform(x_train)
```

```python
x_test_std=std.transform(x_test)
```

# let use DL

```python
from tensorflow.keras.layers import Input, Dense, BatchNormalization, Dropout
from tensorflow.keras.models import Model
from tensorflow.keras.regularizers import L1,L2
```

```python
x_train.info()
```

# use sequential api

```python
from tensorflow.keras import Sequential
```

```python
from tensorflow.keras.callbacks import Callback, EarlyStopping
```

```python
model_seq=Sequential()

model_seq.add(Dense(40,input_dim=44,activation='relu',kernel_regularizer=L2(0.01),name='l1'))
model_seq.add(BatchNormalization(name='bn1'))
model_seq.add(Dropout(0.2,name='d1'))

model_seq.add(Dense(30,activation='relu',kernel_regularizer=L2(0.01),name='l2'))
model_seq.add(BatchNormalization(name='bn2'))
model_seq.add(Dropout(0.2,name='d2'))

model_seq.add(Dense(20,activation='relu',kernel_regularizer=L2(0.01),name='l3'))
model_seq.add(BatchNormalization(name='bn3'))
model_seq.add(Dropout(0.2,name='d3'))

model_seq.add(Dense(10,activation='relu',kernel_regularizer=L2(0.01),name='l4'))
model_seq.add(BatchNormalization(name='bn4'))
model_seq.add(Dropout(0.2,name='d4'))

model_seq.add(Dense(5,activation='relu',kernel_regularizer=L2(0.01),name='l5'))
model_seq.add(BatchNormalization(name='bn5'))
model_seq.add(Dropout(0.2,name='d5'))

model_seq.add(Dense(1,activation='relu',name='output'))
```

```python
model_seq.summary()
```

```python
model_seq.compile(optimizer='adam',loss='poisson',metrics=['mean_absolute_error'])
```

```python
earlystop=EarlyStopping(monitor='loss',patience=5)
```

```python
model_seq.fit(x_train_std,y_train,
          validation_data=(x_test_std,y_test),
             epochs=200,batch_size=120,callbacks=[earlystop])
```
